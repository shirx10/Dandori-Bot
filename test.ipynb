```python
import chromadb
from pipeline import extract_to_dataframe


folder = r"./courses"

def client_setup():
    client = chromadb.Client()
    return client

def collection_setup(client):
    collection_name = "dandori_courses"
    
    if collection_name in client.list_collections():
        collection = client.get_collection(collection_name)
    else:
        collection = client.create_collection(name=collection_name)

    return collection


def embed_data(folder, collection):

    data = extract_to_dataframe(folder)

    documents, metadatas, ids = [], [], []

    for i, item in enumerate(data):
        documents.append(
            f"Course Title: {item['Course_Title']}\n"
            f"Skills: {item['Skills']}\n"
            f"Description: {item['Description']}\n"
            f"Instructor: {item['Instructor']}\n"
            f"Location: {item['Location']}\n"
            f"Cost: {item['Cost']}"
        )
        metadatas.append({
            "Course_Title": item['Course_Title'],
            "Instructor": item['Instructor'],
            "Location": item['Location'],
            "Course_Type": item['Course_Type'],
            "Cost": item['Cost'],
            "Class_ID": item['Class_ID']
        })
        ids.append(f"{i}")

    collection.add(
        documents=documents,
        metadatas=metadatas,
        ids=ids
    )

def query_collection(collection, query, n_results=5):
    results = collection.query(
        query_texts=[query],
        n_results=n_results
    )
    return results
```

```python
client = client_setup()
collection = collection_setup(client)
```

```python
embed_data(folder, collection)
```

```python
query = "I would like a course that offers a relaxed and artistic environment, with a focus on creativity and self-expression. I am interested in learning about various art forms, such as painting, sculpture, and digital art."
```

```python
response = query_collection(collection, query)
```
